#### making performance column for each match on train and test dataset

```python
train_data['team1_roster_ids'] = train_data['team1_roster_ids'].apply(lambda x: list(map(float, x.split(':'))))
train_data['team2_roster_ids'] = train_data['team2_roster_ids'].apply(lambda x: list(map(float, x.split(':'))))

def calculate_team_performance(roster_ids):
    batsman_avg_strike_rate = []
    bowler_avg_economy = []
    bowler_avg_wicket_count=[]

    for player_id in roster_ids:
        if player_id in batsman_data['batsman_id'].values:
            batsman_avg_str_rate = batsman_data.loc[batsman_data['batsman_id'] == player_id, 'avg_strike_rate'].values[0]
            batsman_avg_strike_rate.append(batsman_avg_str_rate)

        if player_id in bowler_data['bowler_id'].values:
            bowler_avg_eco = bowler_data.loc[bowler_data['bowler_id'] == player_id, 'avg_economy'].values[0]
            bowler_avg_wic_count = bowler_data.loc[bowler_data['bowler_id'] == player_id, 'avg_wicket_count'].values[0]
            bowler_avg_economy.append(bowler_avg_eco)
            bowler_avg_wicket_count.append(bowler_avg_wic_count)

    avg_batsman_strike_rate = sum(batsman_avg_strike_rate) / len(batsman_avg_strike_rate) if batsman_avg_strike_rate else 0
    avg_bowler_economy = sum(bowler_avg_economy) / len(bowler_avg_economy) if bowler_avg_economy else 0
    avg_bowler_wicket_count = sum(bowler_avg_wicket_count) / len(bowler_avg_wicket_count) if bowler_avg_wicket_count else 0

    return avg_batsman_strike_rate, avg_bowler_economy, avg_bowler_wicket_count

# Calculate team performance metrics for both team1 and team2
train_data[['team1_avg_strike_rate', 'team1_avg_economy', 'team1_avg_wicket_count']] = train_data['team1_roster_ids'].apply(lambda ids: pd.Series(calculate_team_performance(ids)))
train_data[['team2_avg_strike_rate', 'team2_avg_economy', 'team2_avg_wicket_count']] = train_data['team2_roster_ids'].apply(lambda ids: pd.Series(calculate_team_performance(ids)))
```

```python
# Calculate mean of team1_performance_metric for each match id
team1_str = train_data.groupby('match id')['team1_avg_strike_rate'].mean().reset_index()
team1_eco = train_data.groupby('match id')['team1_avg_economy'].mean().reset_index()
team1_wic = train_data.groupby('match id')['team1_avg_wicket_count'].mean().reset_index()
team1_str.rename(columns={'team1_avg_strike_rate': 'team1_avg_strike_rate_mean'}, inplace=True)
team1_eco.rename(columns={'team1_avg_economy': 'team1_avg_economy_mean'}, inplace=True)
team1_wic.rename(columns={'team1_avg_wicket_count': 'team1_avg_wicket_count_mean'}, inplace=True)

# Merge the means back into test_data based on match id
train_data = pd.merge(train_data, team1_str, on='match id')
train_data = pd.merge(train_data, team1_eco, on='match id')
train_data = pd.merge(train_data, team1_wic, on='match id')
```

```python
# Calculate mean of team2_performance_metric for each match id
team2_str = train_data.groupby('match id')['team2_avg_strike_rate'].mean().reset_index()
team2_eco = train_data.groupby('match id')['team2_avg_economy'].mean().reset_index()
team2_wic = train_data.groupby('match id')['team2_avg_wicket_count'].mean().reset_index()
team2_str.rename(columns={'team2_avg_strike_rate': 'team2_avg_strike_rate_mean'}, inplace=True)
team2_eco.rename(columns={'team2_avg_economy': 'team2_avg_economy_mean'}, inplace=True)
team2_wic.rename(columns={'team2_avg_wicket_count': 'team2_avg_wicket_count_mean'}, inplace=True)

# Merge the means back into test_data based on match id
train_data = pd.merge(train_data, team2_str, on='match id')
train_data = pd.merge(train_data, team2_eco, on='match id')
train_data = pd.merge(train_data, team2_wic, on='match id')
```

##### Selecting relevant columns

```python
# Define feature and target columns
feature_columns = [
    'team_count_50runs_last15', 'team_winp_last5', 'team1_winp_team2_last15',
    'diff_team1_ground', 'win amount', 'diff_strike_rate', 'diff_economy',
    'diff_wicket_count', 'diff_innings_run_rate', 'diff_innings_wickets'
]
target_column = 'winner_01'

# Split features and target from train and co data
X_train = train_data[feature_columns]
y_train = train_data[target_column]

X_cp = cp_data[feature_columns]
y_cp = cp_data[target_column]
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import train_test_split, RandomizedSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler
from scipy.stats import uniform, randint
```

```python
# Standardize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_cp_scaled = scaler.transform(X_cp)
X_combined_scaled = np.vstack((X_train_scaled, X_cp_scaled))
y_combined = np.hstack((y_train, y_cp))

X_test_scaled = scaler.transform(test_data[feature_columns])
```

```python
# Define the parameter distribution for RandomizedSearchCV
param_dist = {
    'n_estimators': randint(50, 300),
    'max_depth': randint(3, 10),
    'learning_rate': uniform(0.01, 0.2),
    'reg_alpha': uniform(0, 1),
    'reg_lambda': uniform(1, 3)
}

# Initialize XGBoost classifier
clf_xgb = XGBClassifier(random_state=42)

# Perform RandomizedSearchCV for hyperparameter tuning
random_search = RandomizedSearchCV(clf_xgb, param_distributions=param_dist,
                                   n_iter=100, cv=5, scoring='accuracy',
                                   verbose=1, random_state=42, n_jobs=-1)
random_search.fit(X_combined_scaled, y_combined)
best_params = random_search.best_params_
print(f"Best parameters: {best_params}")
```

```python
# Initialize the classifier with best parameters
clf_xgb = XGBClassifier(**best_params, random_state=42)

# Perform 5-fold cross-validation to evaluate the model on combined data
cv_scores = cross_val_score(clf_xgb, X_combined_scaled, y_combined, cv=5, scoring='accuracy')
print(f"Cross-validation scores: {cv_scores}")
print(f"Mean cross-validation score: {cv_scores.mean()}")
```

```python
# Fit the model with early stopping on the combined training data
clf_xgb.fit(X_combined_scaled, y_combined,
            eval_set=[(X_combined_scaled, y_combined)],
            eval_metric='logloss',
            early_stopping_rounds=10,
            verbose=False)
```

```python
# Make predictions on training set
train_data['y_pred_01'] = clf_xgb.predict(X_train_scaled)
train_data['win_pred_score'] = clf_xgb.predict_proba(X_train_scaled)[:, 1]
train_data['win_pred_score'] = np.where(train_data['y_pred_01'] == 0, 1 - train_data['win_pred_score'], train_data['win_pred_score'])
train_data['win_pred_team_id'] = np.where(train_data['y_pred_01'] == 0, train_data['team1_id'], train_data['team2_id'])

# Make predictions on test set
test_data['y_pred_01'] = clf_xgb.predict(X_test_scaled)
test_data['win_pred_score'] = clf_xgb.predict_proba(X_test_scaled)[:, 1]
test_data['win_pred_score'] = np.where(test_data['y_pred_01'] == 0, 1 - test_data['win_pred_score'], test_data['win_pred_score'])
test_data['win_pred_team_id'] = np.where(test_data['y_pred_01'] == 0, test_data['team1_id'], test_data['team2_id'])

# Make predictions on cp_data
cp_data['y_pred_01'] = clf_xgb.predict(X_cp_scaled)
cp_data['win_pred_score'] = clf_xgb.predict_proba(X_cp_scaled)[:, 1]
cp_data['win_pred_score'] = np.where(cp_data['y_pred_01'] == 0, 1 - cp_data['win_pred_score'], cp_data['win_pred_score'])
cp_data['win_pred_team_id'] = np.where(cp_data['y_pred_01'] == 0, cp_data['team1_id'], cp_data['team2_id'])
```
